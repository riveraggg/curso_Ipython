El movimiento planetario está dado por la ley de gravitación universal:

$$ m_p \frac{d^2 \vec r}{dt^2} =  - GM_s m_p \frac{\vec r}{|r|^3}$$

El movimiento no depende de la masa del planeta, $m_s$, podemos hacer el cambio de variables:
$T= \sqrt{ \tfrac{GM_s}{R_T^3} }t$, $\vec R = \tfrac{\vec r}{R_T}$ y la ecuación anterior queda de la siguiente manera:
$$ \frac{d^2\vec R}{dT^2} = - \frac{\vec R}{|R|^3}$$

* $G = 6.674 \times 10^{-11} \tfrac{Nm²}{kg^2}$

* $M_s = 1.9891 \times 10^{30} kg$

* $R_T = 1.52 \times 10^{11} m$

Librerias de *IPython* que utilizaremos:

```python
from pylab import *
from scipy.integrate import  *
%matplotlib inline
```

```python
G = 6.674e-11
M_s = 1.9891e30
R_T = 1.52e11
Cte = sqrt((G*M_s)/(R_T**3))
yr = 60*60*24*365
P=(2*pi)/Cte
```

```python
def gravitacion(r_v,t):
    # r_v contendra las como vector las variables (x,y,vx,vy)
    r = sqrt(r_v[0]**2+r_v[1]**2)
    dx = r_v[2]
    dy = r_v[3]
    dvx = -r_v[0]/r**3
    dvy = -r_v[1]/r**3
    return [dx,dy,dvx,dvy]
```

```python
tiempo = linspace(0,2*pi)
sol = odeint(gravitacion,[1,0,0,1],tiempo)
x,y = sol[:,0], sol[:,1]
xlim([-1.1,1.1])
ylim([-1.1,1.1])
axes().set_aspect('equal')
grid(True)
plot ([0],[0],'o')
plot (x,y,'o')
show()
```

```python
tiempo = linspace(0,2*pi)
vx = 0.5
vy = sqrt(1-vx**2)
sol = odeint(gravitacion,[1,0,vx,vy],tiempo)
x,y = sol[:,0], sol[:,1]
xlim([-1.5,1.5])
ylim([-1.5,1.5])
axes().set_aspect('equal')
plot([0],[0],'o')
grid(True)
plot (x,y,'o')
show()
```
